```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv('/content/diabetes.csv')
```

Splitting Data

```python
X = Data.drop(['Outcome'],axis=1)
Y = Data['Outcome']
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, stratify=Y, test_size=0.2,random_state=42)
```

Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_Scaled = scaler.fit_transform(X_train)
X_test_Scaled = scaler.transform(X_test)

```

Model Training

SVM Classifier

```python
from sklearn import svm
```

```python
SVM = svm.SVC(kernel='linear')
```

```python
SVM.fit(X_train_Scaled, y_train)
```

Prediction

```python
train_data_prediction = SVM.predict(X_train_Scaled)
test_data_prediction = SVM.predict(X_test_Scaled)
```

Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
Train_Acc = accuracy_score(y_train,train_data_prediction)

Test_Acc = accuracy_score(y_test,test_data_prediction)
```

```python
print(Train_Acc)
print(Test_Acc)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
Log_Reg = LogisticRegression()
```

```python
Log_Reg.fit(X_train_Scaled,y_train)
```

```python
Log_train_data_prediction = Log_Reg.predict(X_train_Scaled)
Log_test_data_prediction = Log_Reg.predict(X_test_Scaled)
```

```python
Log_Train_Acc = accuracy_score(y_train,Log_train_data_prediction)

Log_Test_Acc = accuracy_score(y_test,Log_test_data_prediction)
```

```python
print(Log_Train_Acc)
print(Log_Test_Acc)
```
